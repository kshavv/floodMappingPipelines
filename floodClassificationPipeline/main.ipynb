```python


from flood_mapping import (
    download_temporal_images_for_years,
    classify_years,
)

# ── Config ────────────────────────────────────────────────────
STATE       = 'Kerala'
YEARS       = [2019, 2020, 2021, 2022, 2023, 2024]
ASSET_ROOT  = 'projects/gentle-operator-308420/assets/TemporalImagesNew'
TRAINING_FC = 'projects/gentle-operator-308420/assets/finalDataset/full_dataset_v3'


# ── Step 1: download (= export to EE asset) the full-year temporal stacks ──
# One 27-band asset per year: RF_water_FullYear_<year>_kerala.
# skip_existing=True (default) makes this idempotent — already-exported
# years are skipped automatically.
download_temporal_images_for_years(
    state=STATE,
    years=YEARS,
    training_fc=TRAINING_FC,
    asset_root=ASSET_ROOT,
    destinations='asset',
)

# At this point, open the Earth Engine Tasks tab and let all 6 export
# tasks finish before running Step 2. They take hours for a state-sized
# ROI. Re-running Step 1 after they finish is safe (just prints "already
# exists, skipping").


```

```python
# ── Step 2: classify each year into a 27-band Classified image ──
# One classification asset per year: Classified_<year>_kerala.
# Reads the full-year stacks Step 1 produced and writes state-shaped,
# polygon-masked output (no rectangle bounding boxes).

OUTPUT_ROOT = 'projects/gentle-operator-308420/assets/Classified'
ASSET_ROOT  = 'projects/gentle-operator-308420/assets/TemporalImagesNew'
STATE       = 'Kerala'
YEARS       = [2024]

classify_years(
    state=STATE,
    years=YEARS,
    asset_root=ASSET_ROOT,
    destinations='asset',
    output_root=OUTPUT_ROOT,
)
```
